## **Question 2 : Explain the difference between List and Tuple in Python.**

*Ans* - **List** :
*    **Definition**- List is an ordered collection of item, where each items can be of any data type ( eg. Interger or Strings. ), a list is written in using square brackets [ ]
*    **Mutation**- Mutaion can be possible inside a list that means items can be added, removed or changed.
*    **Indexed**- Any items inside the list can be accessed using their position in the list.
*    **Duplicacy**- Items inside a list can have duplicate values.
       
       **Tuple**:
*    **Definition**- Tuple is a type of data in which mutaion is not possible, that means items inside a tuple can't be added, removed or modified. A Tuple can be written using ( ) round brackets.
*    **Usage**-it is manly used when you need to have a fixed collection of items which won't change. for e.g. co-ordinates, configuration setting or dictionary keys.
*    **Indexed**- Like List indexing is also possible inside a Tuple as the items are fixed in their position.
*    **Duplicacy**- Items inside a Tuple can have duplicate values like List.
