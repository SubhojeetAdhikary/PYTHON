## **Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure.**

```python
import pandas as pd
import numpy as np
```

```python
#Ans1

# Loading retail data

df = pd.read_csv("/retail_dataset .csv")
df
```

```python
# First 5 rows

df.head()
```

```python
# Dataset information to understand its structure.

df.info()
df.shape
```

## **Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.**

```python
# Ans2
# Identify missing values in each column

missing_values = df.isnull().sum()
missing_values
```

```python
# The total number of missing values present in the dataset.

total_missing_values = missing_values.sum()
total_missing_values
```

## **Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.**

```python
# Ans3
# Replacing missing values in the Quantity and Price columns

df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

```python
# Checking whether missing values are still present or not.

print(df[['Quantity', 'Price']].isnull().sum())
```

## **Q4. Remove all rows where the Product Category or Region column contains missing values.**

```python
#Ans3
# Removing all rows where the Product Category or Region column contains missing values

df = df.dropna(subset=['Product Category', 'Region'])
```

```python
# Checking whether those columns are removed or not.

df.shape
```

## **Q5. Create a new column called Revenue using NumPy, where:**
- Revenue = Quantity × Price
- Display the updated dataset with the new column.

```python
# Ans5
# Creating a new column called Revenue

df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

# Displaying the updated dataset with the new column.

df.head()
```

## **Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.**

```python
# Ans6

# Calculating the total revenue generated from all transactions in the dataset.

total_revenue = np.sum(df['Revenue'])
total_revenue
```

## **Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.**

```python
# Ans7

# The total revenue generated by each category.

category_revenue = df.groupby('Product Category')['Revenue'].sum()
category_revenue
```

## **Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.**

```python
# Ans8

# Top 3 categories with the highest revenue.

top_3_categories = category_revenue.nlargest(3)
top_3_categories
```

```python
# Bottom 3 categories with the lowest revenue.

bottom_3_categories = category_revenue.nsmallest(3)
bottom_3_categories
```

## **Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.**

```python
# Ans9

# Grouping the dataset by Region and calculate the total revenue generated in each region.

region_revenue = df.groupby('Region')['Revenue'].sum()
region_revenue
```

```python
## Identify the region with the highest revenue and the region with the lowest revenue.

highest_revenue_region = region_revenue.idxmax()
lowest_revenue_region = region_revenue.idxmin()

print("Region with the highest revenue:", highest_revenue_region)
print("Region with the lowest revenue:", lowest_revenue_region)
```

## **Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate thetotal monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.**

```python
#Ans10

# First convert Date column into Date time.

df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)

# Extracting the month from the Date column using Pandas.

df['Month'] = df['Date'].dt.month_name()

# Grouping the dataset by month to calculate the total monthly revenue.

monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_revenue
```

```python
# Calculating the mean, median, and standard deviation of the Revenue column using NumPy.

mean_revenue = np.mean(df['Revenue'])
median_revenue = np.median(df['Revenue'])
std_revenue = np.std(df['Revenue'])

print("Mean Revenue:", mean_revenue)
print("Median Revenue:", median_revenue)
print("Standard Deviation of Revenue:", std_revenue)
```
