## **Q3 Customer Purchase Analysis**

Identify the top 5 customers based on total sales.

Hint: Use GroupBy with aggregation.

```python
# Ans 3
# fist find which customer contribute how much in total sales.

customer_sales = df.groupby('Customer')['Sales'].sum()
customer_sales
```

```python
# Now find top 5 customer contributing most in sales

top_5_customers = customer_sales.nlargest(5)
top_5_customers
```

## **Q4 Regional Sales Performance**

Calculate the total sales generated by each region.

Then identify:

Which region generated the highest revenue?

```python
# Ans 4
# Total sales generated by each region.

region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
# Region generated the highest revenue

highest_revenue_region = region_sales.idxmax()
highest_revenue_region
```

## **Q5 Product Performance Analysis**

Calculate the total quantity sold for each product.

Which product appears to be the most popular?



```python
# Ans 5
# The total quantity sold for each product

quantity_sold = df.groupby('Product')['Quantity'].sum()
quantity_sold
```

```python
# Which product appears to be the most popular?

most_popular_product = quantity_sold.idxmax()
most_popular_product
```

## **Q6 Multi-Level Grouping**

## **Q9 Discount Impact Study**

**Analyze the relationship between discount and profit.**

Tasks:

Calculate the average profit for each discount level

Identify whether higher discounts lead to lower profit



```python
# Ans 9
# Calculate the average profit for each discount level

avg_profit_discount = df.groupby('Discount')['Profit'].mean()
avg_profit_discount
```

```python
# Identify whether higher discounts lead to lower profit

print(avg_profit_discount.idxmin(),'-', avg_profit_discount.min())
```

As per the above analysis we can see that when we are increasing discounts our profits falls significantly in each of the discount levels.

## **Q10 End-to-End Data Wrangling Task**

**Perform the following workflow:**

- Filter the dataset where Sales > 3000

- Group data by Region and Product

- Calculate total sales

- Reshape the result into a pivot table

Explain the business insights obtained from this analysis.

```python
# Ans 10
# Filter the dataset where Sales > 3000

df2 = df[df['Sales']> 3000]
df2
```

```python
# Group data by Region and Product

df2.groupby(['Region','Product'])['Sales'].sum().reset_index()
```

```python
# Calculate total sales

df2.groupby('Region')['Sales'].sum()
```

```python
# Reshape the result into a pivot table

df3 = pd.pivot_table(df2, index= 'Region', columns= 'Product', values= 'Sales', aggfunc= 'sum')
df3
```
